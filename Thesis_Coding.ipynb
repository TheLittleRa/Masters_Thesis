## Begin The Preprocessing

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(df[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']])

# Convert back to DataFrame for compatibility
scaled_df = pd.DataFrame(scaled_data, columns=['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume'])
scaled_df
```

## Creating The Sequences

```python

def create_multi_output_sequences(data, seq_len=25, forecast_horizon=5):
    X, y = [], []
    for i in range(len(data) - seq_len - forecast_horizon + 1):
         # Extract a window of 'sequence_length' days of features
        X.append(data.iloc[i:i + seq_len].values)
        # Use the 'Close' price of the day following the window as the target
        y.append(data.iloc[i + seq_len:i + seq_len + forecast_horizon]['Close'].values)
    return np.array(X), np.array(y)
# Generate input sequences and corresponding target values
X, y = create_multi_output_sequences(scaled_df, 25, 5)


```

# Start Building Training Process

```python
# Determine the index at which to split the data into training and testing sets (85% training)
split_index = int(len(X) * 0.85)

# Split the input sequences into training and testing sets
X_train, X_test = X[:split_index], X[split_index:]

# Split the corresponding target values into training and testing sets
y_train, y_test = y[:split_index], y[split_index:]
```

## Building The LSTM Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.optimizers import Adam

model = Sequential()
model.add(LSTM(256, return_sequences=True, input_shape=(30, 6)))
model.add(Dropout(0.3))
model.add(LSTM(256, return_sequences=True))
model.add(Dropout(0.3))
model.add(LSTM(128))
model.add(Dropout(0.2))
model.add(Dense(64, activation='relu'))  
model.add(Dropout(0.2))
model.add(Dense(5)) # Predict 5 days

model.compile(optimizer= Adam(learning_rate=0.001), loss='mean_squared_error')


from tensorflow.keras.callbacks import EarlyStopping

# Define the EarlyStopping callback
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Train the model with EarlyStopping callback
model.fit(
    X_train, y_train, 
    epochs=100,                # We can train for more epochs, EarlyStopping will handle when to stop
    batch_size=32, 
    validation_split=0.2,      # 20% of the data will be used for validation
    callbacks=[early_stopping] # Add the EarlyStopping callback here
)


```
